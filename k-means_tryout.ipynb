```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
image = cv2.imread('Data/training_set/yes/Y24.jpg')
image = cv2.resize(image, (256, 256))
im = image.copy()
new = image.copy()
cv2.imshow("Original",image)
cv2.waitKey(0);
gray=image.copy()
gray=cv2.cvtColor(gray,cv2.COLOR_BGR2RGB)
gray=cv2.cvtColor(gray,cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray",gray)
cv2.waitKey(0);
biggest = np.amax(gray)
print("Biggest")
print(biggest)
blur=cv2.GaussianBlur(gray,(5,5),0)
cv2.imshow("Blur",gray)
cv2.waitKey(0);
pixels=blur.reshape((-1,2))
pixels=np.float32(pixels)
k = 4
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0) #used to stop when req accuracy or iteration is reached
retval,labels,centers = cv2.kmeans(pixels,k,None,criteria,100,cv2.KMEANS_RANDOM_CENTERS)
labels = labels.flatten()
centers = np.uint8(centers)
seg_image = centers[labels.flatten()]
seg_image = seg_image.reshape((gray.shape))
cv2.imshow("Segmented",seg_image)
cv2.waitKey(0);

contours, hierarchy = cv2.findContours(seg_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
contours = sorted(contours,key = cv2.contourArea)
area = 0
cntmin=0
it=image
im = image
imin = image
imax=image
mask = np.copy(gray)
mask = mask.reshape((-1,2))
mask[labels == 1] = [0, 0]
mask[labels == 2] = [0, 0]
mask[labels == 3] = [0, 0]
mask = mask.reshape(gray.shape)
cv2.imshow("mask 0",mask)
cv2.waitKey(0);
ret,mask1 = cv2.threshold(mask,250,255,cv2.THRESH_BINARY)
#kernel = np.ones((4,4))
#cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4))
#erode = cv2.erode(mask1,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
#erode = cv2.dilate(erode,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
cnt = (cv2.countNonZero(mask1))
if cnt>=cntmin:
    cntmin = cnt
    #imin = erode
    it=mask
print("mask 0")
print(cnt)
im = image
mask = np.copy(gray)
mask = mask.reshape((-1,2))
print("1")
mask[labels == 0] = [0, 0]
mask[labels == 2] = [0, 0]
mask[labels == 3] = [0, 0]
mask = mask.reshape(gray.shape)
cv2.imshow("mask 1",mask)
cv2.waitKey(0);
ret,mask1 = cv2.threshold(mask,250,255,cv2.THRESH_BINARY)
#erode = cv2.erode(mask1,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
#erode = cv2.dilate(erode,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
cnt = (cv2.countNonZero(mask1))
if cnt>=cntmin:
    cntmin = cnt
    #imin = erode
    it=mask
print("mask 1")
print(cnt)
im = image
mask = np.copy(gray)
mask = mask.reshape((-1,2))
print("2")
mask[labels == 0] = [0, 0]
mask[labels == 1] = [0, 0]
mask[labels == 3] = [0, 0]
mask = mask.reshape(gray.shape)
cv2.imshow("mask 2",mask)
cv2.waitKey(0);
ret,mask1 = cv2.threshold(mask,250,255,cv2.THRESH_BINARY)
#erode = cv2.erode(mask1,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
#erode = cv2.dilate(erode,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
cnt = (cv2.countNonZero(mask1))
if cnt>=cntmin:
    cntmin = cnt
    #imin = erode
    it=mask
print("mask 2")
print(cnt)
im = image
mask = np.copy(gray)
mask = mask.reshape((-1,2))
print("3")
mask[labels == 0] = [0, 0]
mask[labels == 1] = [0, 0]
mask[labels == 2] = [0, 0]
mask = mask.reshape(gray.shape)
cv2.imshow("mask 3",mask)
cv2.waitKey(0);
ret,mask1 = cv2.threshold(mask,250,255,cv2.THRESH_BINARY)
#erode = cv2.erode(mask1,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
#erode = cv2.dilate(erode,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
cnt = (cv2.countNonZero(mask1))
if cnt>=cntmin:
    cntmin = cnt
    #imin = erode
    it=mask
print("mask 3")
print(cnt)
print("cntmin is")
print(cntmin)
if cntmin==0:
    tumor=image
    cv2.imshow("Tumor part",image)
    cv2.waitKey(0);
else:
    cv2.imshow("Result mask is",it)
    cv2.waitKey(0);
    sx = cv2.Sobel(it,cv2.CV_32F,1,0)

    sy = cv2.Sobel(it,cv2.CV_32F,0,1)

    m = cv2.magnitude(sx,sy)

    m = cv2.normalize(m,None,0.,255.,cv2.NORM_MINMAX,cv2.CV_8U)
    #cv2.imshow("Sobel",m)
    #cv2.waitKey(0);

    contours, hierarchy = cv2.findContours(m, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
    contours = sorted(contours,key = cv2.contourArea)
    x=0
    y=0
    w=256
    h=256
    print(len(contours))
    s="cont"
    for i in range(len(contours)-1,-1,-1):
        dupe = image.copy()
        x,y,w,h = cv2.boundingRect(contours[i])
        g=s+str(i)
        dupe = cv2.rectangle(dupe,(x,y),(x+w,y+h),(0,255,0),3)
        cv2.imshow(g,dupe)
        cv2.waitKey(0);
        if w*h<(22000):
            break
    tumor = im[y:y+h,x:x+w]
    cv2.imshow("Tumor Part",tumor)
    cv2.waitKey(0);
    im = cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),3)
    cv2.imshow("Rect",im)
    cv2.waitKey(0);
cv2.destroyAllWindows();
```
