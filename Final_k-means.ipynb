```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
def write_image(image,tofolder,cnt):
    to_folder = tofolder+'NO'+str(cnt)+'.jpg'
    cv2.imwrite(to_folder,image)
    cv2.waitKey(0);
    cnt=cnt+1
def kmeans(folder,tofolder):
    cnt1=0
    for file in os.listdir(folder):
        cnt1=cnt1+1
        image = cv2.imread(os.path.join(folder,file))
        wf,hf,df = image.shape
        ##############    CODE STARTS HERE  #########################
        #original = image
        image = cv2.resize(image, (256, 256))
        im = image
        new = image.copy()
        gray=image.copy()
        gray=cv2.cvtColor(gray,cv2.COLOR_BGR2RGB)
        gray=cv2.cvtColor(gray,cv2.COLOR_BGR2GRAY)
        biggest = np.amax(gray)
        #gray = cv2.equalizeHist(gray)
        blur=cv2.GaussianBlur(gray,(5,5),0)
        #blur = cv2.equalizeHist(blur)
        pixels=blur.reshape((-1,2))
        pixels=np.float32(pixels)
        k = 4
        criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
        retval,labels,centers = cv2.kmeans(pixels,k,None,criteria,100,cv2.KMEANS_RANDOM_CENTERS)
        labels = labels.flatten()
        centers = np.uint8(centers)
        seg_image = centers[labels.flatten()]
        seg_image = seg_image.reshape((gray.shape))


        sx = cv2.Sobel(seg_image,cv2.CV_32F,1,0)

        sy = cv2.Sobel(seg_image,cv2.CV_32F,0,1)

        m = cv2.magnitude(sx,sy)

        m = cv2.normalize(m,None,0.,255.,cv2.NORM_MINMAX,cv2.CV_8U)

        contours, hierarchy = cv2.findContours(seg_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
        contours = sorted(contours,key = cv2.contourArea)
        area = 0
        cntmin=0
        it=image
        im = image
        imin = image
        imax=image
        mask = np.copy(gray)
        mask = mask.reshape((-1,2))
        mask[labels == 1] = [0, 0]
        mask[labels == 2] = [0, 0]
        mask[labels == 3] = [0, 0]
        mask = mask.reshape(gray.shape)
        ret,mask1 = cv2.threshold(mask,250,255,cv2.THRESH_BINARY)
        #kernel = np.ones((4,4))
        #cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4))
        erode = cv2.erode(mask1,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
        erode = cv2.dilate(erode,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
        cnt = (cv2.countNonZero(mask1))
        if cnt>=cntmin:
            cntmin = cnt
            imin = erode
            it=mask
        im = image
        mask = np.copy(gray)
        mask = mask.reshape((-1,2))
        mask[labels == 0] = [0, 0]
        mask[labels == 2] = [0, 0]
        mask[labels == 3] = [0, 0]
        mask = mask.reshape(gray.shape)
        ret,mask1 = cv2.threshold(mask,250,255,cv2.THRESH_BINARY)
        erode = cv2.erode(mask1,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
        erode = cv2.dilate(erode,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
        cnt = (cv2.countNonZero(mask1))
        if cnt>=cntmin:
            cntmin = cnt
            imin = erode
            it=mask
        im = image
        mask = np.copy(gray)
        mask = mask.reshape((-1,2))
        mask[labels == 0] = [0, 0]
        mask[labels == 1] = [0, 0]
        mask[labels == 3] = [0, 0]
        mask = mask.reshape(gray.shape)
        ret,mask1 = cv2.threshold(mask,250,255,cv2.THRESH_BINARY)
        erode = cv2.erode(mask1,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
        erode = cv2.dilate(erode,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
        cnt = (cv2.countNonZero(mask1))
        if cnt>=cntmin:
            cntmin = cnt
            imin = erode
            it=mask
        im = image
        mask = np.copy(gray)
        mask = mask.reshape((-1,2))
        mask[labels == 0] = [0, 0]
        mask[labels == 1] = [0, 0]
        mask[labels == 2] = [0, 0]
        mask = mask.reshape(gray.shape)
        ret,mask1 = cv2.threshold(mask,250,255,cv2.THRESH_BINARY)
        erode = cv2.erode(mask1,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
        erode = cv2.dilate(erode,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4)),iterations = 1)
        cnt = (cv2.countNonZero(mask1))
        if cnt>=cntmin:
            cntmin = cnt
            imin = erode
            it=mask
        if cntmin==0:
            tumor=image
        else:
            sx = cv2.Sobel(it,cv2.CV_32F,1,0)

            sy = cv2.Sobel(it,cv2.CV_32F,0,1)

            m = cv2.magnitude(sx,sy)

            m = cv2.normalize(m,None,0.,255.,cv2.NORM_MINMAX,cv2.CV_8U)

            contours, hierarchy = cv2.findContours(m, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
            contours = sorted(contours,key = cv2.contourArea)
            x=0
            y=0
            w=256
            h=256
            for i in range(len(contours)-1,-1,-1):
                dupe = image.copy()
                x,y,w,h = cv2.boundingRect(contours[i])
                if w*h<(22000):
                    break
            tumor = im[y:y+h,x:x+w]
        tumor = cv2.resize(tumor,(wf,hf))
        write_image(tumor,tofolder,cnt1)
#kmeans('dataset/training_set/yes/','Try/training_set/yes/')
#kmeans('dataset/training_set/no/','Try/training_set/no/')
#kmeans('dataset/test_set/yes/','Try/test_set/yes/')
#kmeans('dataset/test_set/no/','Try/test_set/no/')
kmeans('dataset/single_prediction/','Try/single_prediction/')
```
